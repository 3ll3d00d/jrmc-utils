```python
import pandas as pd 

pd.options.display.max_columns = None
pd.set_option('display.width',500)
pd.set_option('display.max_rows',1200)

df = pd.read_csv('/media/films/crop_analysis.csv').drop_duplicates(subset=['name','res'])
m_df = df[df['exact'] == True]
df = df[df['exact'] == False].drop(columns='exact')
f"Exact: {len(m_df)} Remainder: {len(df)}" 
```

```python
m_crop = df['mc_crop'].str.split('x', expand=True).astype(int)
df = df.assign(**{
    'mc_w': m_crop[2] - m_crop[0], 
    'mc_h': m_crop[3] - m_crop[1],
    'mc_ar': round((m_crop[2] - m_crop[0]) / (m_crop[3] - m_crop[1]), 2)
})
old_cols = df.columns.tolist()
new_cols = ['name', 'res'] + [c for c in old_cols if c.startswith('mc_')] + [c for c in old_cols if c.endswith('fcrop')]
df = df[new_cols]
df
```

```python

w_tolerance_px = 16
h_tolerance_px = 16

new_cols = ['name', 'res', 'mc_crop', 'mc_ar']

for limit in ['18', '25']:
    for skip in ['24', '8']:
        suffix = f'{limit}_{skip}_'
        f_crop = df[f'{suffix}fcrop'].str.split('x', expand=True).astype(int)
        df = df.assign(**{
            f'{suffix}fw': f_crop[2] - f_crop[0],
            f'{suffix}fh': f_crop[3] - f_crop[1],
            f'{suffix}f_ar': round((f_crop[2] - f_crop[0]) / (f_crop[3] - f_crop[1]), 2),
        })
        df = df.assign(**{
            f'{suffix}w_delta': df.apply(lambda r: abs((r[f'{suffix}fw'] - r['mc_w'])), axis=1).values,
            f'{suffix}h_delta': df.apply(lambda r: abs((r[f'{suffix}fh'] - r['mc_h'])), axis=1).values
        })
        df = df.assign(**{
            f'{suffix}w_ok': df.apply(lambda r: r[f'{suffix}w_delta'] <= w_tolerance_px, axis=1).values,
            f'{suffix}h_ok': df.apply(lambda r: r[f'{suffix}h_delta'] <= h_tolerance_px, axis=1).values,
        })
        df[f'{suffix}ok'] = df.apply(lambda row: row[f'{suffix}w_ok'] and row[f'{suffix}h_ok'], axis=1) 

        new_cols.extend([f'{suffix}{v}' for v in ['ok', 'fw', 'fh', 'f_ar', 'w_ok', 'h_ok', 'w_delta', 'h_delta', 'fcrop']])

df = df[new_cols]
df
```
